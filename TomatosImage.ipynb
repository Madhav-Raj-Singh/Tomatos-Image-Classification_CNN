```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
from tensorflow import keras
```

```python
##generators, gives numpy array
train_ds=keras.utils.image_dataset_from_directory(
    directory='/content/drive/MyDrive/content/ieee-mbl-cls/train',
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(128, 128)
)
val_ds=keras.utils.image_dataset_from_directory(
    directory='/content/drive/MyDrive/content/ieee-mbl-cls/val',
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(128, 128)
)
```

```python
##Normalize each pixel between 0 to 1
def process(image,label):
  image=tf.cast(image/128. ,tf.float32)
  return image,label

train_ds=train_ds.map(process)
val_ds=val_ds.map(process)
```

```python
##CNN model
classifier=Sequential()

classifier.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(128,128,3)))
classifier.add(MaxPooling2D(pool_size=(3,3),strides=2,padding='valid'))

classifier.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
classifier.add(MaxPooling2D(pool_size=(3,3),strides=2,padding='valid'))

classifier.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
classifier.add(MaxPooling2D(pool_size=(3,3),strides=2,padding='valid'))

classifier.add(Flatten())



classifier.add(Dense(128,activation='relu'))
classifier.add(Dense(64,activation='relu'))
classifier.add(Dense(4,activation='softmax'))
```

```python
classifier.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow as tf
es=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.3,
    patience=20,
    verbose=3,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
classifier.fit(train_ds,validation_data=val_ds,batch_size=30,epochs=10,callbacks=es)
```
